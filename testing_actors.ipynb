```python
# import modules
import sys

try:
    sys.path.append('/home/karenli/carla/PythonAPI/carla/dist/carla-0.9.14-py3.8-linux-x86_64.egg')
except IndexError:
    print('Error: CARLA PythonAPI not found.')
    pass

import carla

import time
```

```python

# connect to the client and get the world object
client = carla.Client('localhost', 2000)
world = client.get_world()
```

```python
# get pedestrian blueprint
blueprint_library = world.get_blueprint_library()
pedestrian_bp = blueprint_library.find('walker.pedestrian.0001')
```

```python
# spawn pedestrian
transform = carla.Transform(carla.Location(x=9210.0, y=21570.0, z=483.0), carla.Rotation(yaw=180))
actor = world.spawn_actor(pedestrian_bp, transform)
```

```python
# move the spectator to the pedestrian
spectator = world.get_spectator()
transform = carla.Transform(actor.get_location(), actor.get_transform().rotation)
spectator.set_transform(transform)
```
